```python
import pandas as pd
import numpy as np
import random

output_1 = pd.read_csv('output (1).csv')
output_2 = pd.read_csv('output (2).csv')
output_3 = pd.read_csv('output (3).csv')

valid_datas = len(output_1)
classes = 42
predict = []

for i in range(valid_datas) :
    arr = np.zeros(42)
    
    arr[output_1.iloc[i].item()] += 1
    arr[output_2.iloc[i].item()] += 1
    arr[output_3.iloc[i].item()] += 1
    
    max_index = np.where(arr == arr.max())
    print(max_index)
    
    if len(max_index[0]) == 1 :
        if output_1.iloc[i].item() != max_index[0][0] :
            print('2, 3 => ', max_index[0][0])
        predict.append(max_index[0][0])
    else : 
        if output_1.iloc[i].item() in max_index[0] :
            print('1 => : ', output_1.iloc[i].item())
            predict.append(output_1.iloc[i].item())
        elif output_2.iloc[i].item() in max_index[0] :
            print('2 => : ', output_1.iloc[i].item())
            predict.append(output_2.iloc[i].item())
        else :
            c = random.choice(max_index[0])
            print(i, max_index[0], ' : ', c, arr[c])
            predict.append(arr[c])
        
out = pd.DataFrame(predict, columns = ['pred'])
out.to_csv('hard_vote_over_74.csv', index = False)
```
