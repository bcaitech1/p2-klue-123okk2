```python
import pandas as pd

train_df = pd.read_csv('/opt/ml/input/data/train/train_aug_fix.tsv', sep='\t', header=None)
test_df = pd.read_csv('/opt/ml/input/data/test/test.tsv', sep='\t', header=None)

def insert_ent_tag(text, ent1_start_idx, ent1_end_idx, ent2_start_idx, ent2_end_idx):
    new_text = ''
    
    for i, ch in enumerate(text):
        if i == ent1_start_idx:
            new_text += '<E1>'
            new_text += ch
            
            if ent1_start_idx == ent1_end_idx:
                new_text += '</E1>'
                
        elif i == ent1_end_idx:
            new_text += ch
            new_text += '<E2>'
            
        elif i == ent2_start_idx:
            new_text += '</E2>'
            new_text += ch
            
            if ent2_start_idx == ent2_end_idx:
                new_text += '[/E]'
                
        elif i == ent2_end_idx:
            new_text += ch
            new_text += '[E]'
            
        else:
            new_text += ch
    
    # entity index 위치 변경
    if ent1_start_idx < ent2_start_idx:
        ent1_end_idx += 2
        ent2_start_idx += 2
        ent2_end_idx += 4
    else:
        ent2_end_idx += 2
        ent1_start_idx += 2
        ent1_end_idx += 4

    return new_text, ent1_start_idx, ent1_end_idx, ent2_start_idx, ent2_end_idx


train_df[[1, 3, 4, 6, 7]] = train_df.apply(lambda x: insert_ent_tag(x[1], int(x[3]), int(x[4]), int(x[6]), int(x[7])), axis=1, result_type='expand')
test_df[[1, 3, 4, 6, 7]] = test_df.apply(lambda x: insert_ent_tag(x[1], int(x[3]), int(x[4]), int(x[6]), int(x[7])), axis=1, result_type='expand')


# csv로 저장
train_df.to_csv('/opt/ml/input/data/train/one_labeled_train_aug_fix.tsv', index=False, sep='\t', header=None)
test_df.to_csv('/opt/ml/input/data/test/one_labeled_test.tsv', index=False, sep='\t', header=None)
```
